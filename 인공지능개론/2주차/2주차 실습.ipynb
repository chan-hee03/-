```python
import pandas as pd

#데이터 로딩
df=pd.read_csv(r"C:\Users\김찬희\인공지능개론\2주차\titanic.csv")

### 데이터 전처리 
#결측치 확인
print(df.isnull().sum())
#결측치 age: 177, Cabin: 687,Embarked: 2


# 레이블 갯수 확인 549/342
print(df["Survived"].value_counts())


#결측치 제거  (Cabin 열 삭제)
df=df.drop(columns=["Cabin"])

#결측치 제거 (age 평균으로 채우기)
mean_Age=df["Age"].mean()
df["Age"] = df["Age"].fillna(mean_Age)

# 결측치 제거 (Embarked 결측 샘플 삭제)
df = df.dropna()


# 레이블 갯수 재확인 549/340
print(df["Survived"].value_counts())


# 불필요 컬럼 삭제 (승객 ID, 이름, 티켓번호)
df = df.drop(columns=["PassengerId", "Name", "Ticket"])


print(df.columns)
# 인코딩 (성별, 탑승항구)
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
columns_to_encode = ["Sex", "Embarked"]
for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])

print(df.head)




### 인공지능 학습

# 라벨과 특성 분류
x=df.drop("Survived",axis=1)
y=df["Survived"]

# train-test 분류
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)

# 모델 import
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# 학습
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
DT=DecisionTreeClassifier()
DT.fit(x_train,y_train)
LR=RandomForestClassifier()
LR.fit(x_train,y_train)
SVM=RandomForestClassifier()
SVM.fit(x_train,y_train)
Kn=KNeighborsClassifier()
Kn.fit(x_train,y_train)

# 예측
RF_pred=RF.predict(x_test)
DT_pred=DT.predict(x_test)
LR_pred=LR.predict(x_test)
SVM_pred=SVM.predict(x_test)
Kn_pred=Kn.predict(x_test)




# accuracy 확인
from sklearn.metrics import accuracy_score
print(f"RF 정확도 : {accuracy_score(y_test, RF_pred)}")
print(f"DT 정확도 : {accuracy_score(y_test, DT_pred)}")
print(f"LR 정확도 : {accuracy_score(y_test, LR_pred)}")
print(f"SVM 정확도 : {accuracy_score(y_test, SVM_pred)}")
print(f"Kn 정확도 : {accuracy_score(y_test, Kn_pred)}")


# confusion matrix 확인
from sklearn.metrics import confusion_matrix
print(f"RF Confusion Matrix : {accuracy_score(y_test,RF_pred)}")
print(f"DT Confusion Matrix : {accuracy_score(y_test,DT_pred)}")
print(f"LR Confusion Matrix : {accuracy_score(y_test,LR_pred)}")
print(f"SVM Confusion Matrix : {accuracy_score(y_test,SVM_pred)}")
print(f"Kn Confusion Matrix : {accuracy_score(y_test,Kn_pred)}")


################ 자투리


# Pclass에 대한 비용 평균 구하기

# df['Pclass'] == n  ←  Pclass가 n 이면 True, 아니면 False로 이루어진 불리언시리즈 생성
# df[불리언시리즈]   ←    불리언 시리즈가 True인 행만 남기게됨
print(df[df["Pclass"] == 3]["Fare"].mean())
print(df[df["Pclass"] == 2]["Fare"].mean())
print(df[df["Pclass"] == 1]["Fare"].mean())

# df[]==k 를 이용한 인코딩?
# df["Sex"] = df["Sex"] == "male"  #쌉가능..?

# 탑승항구별 Pclass 비율  *원형그래프 사용해보기


# sklearn.metrics.accuracy_score(실제값, 예측값) ↔ (y_test, predict(x_test)) 
# score는 (특성, 타겟) ↔  (x_test, y_test)
# 인자순서가 score와 다르게 반대인 이유는 함수의 설계철학에 의한...  ??? 



```
